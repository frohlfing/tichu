# Datenanalyse der Logdateien von Brettspielwelt

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

data = np.random.normal(0, 1, 100)
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q-Plot")
plt.grid(True)
plt.show()
```

```python
plt.hist(data, bins=30, edgecolor='black')
plt.title("Histogramm")
plt.xlabel("Wert")
plt.ylabel("Häufigkeit")
plt.grid(True)
plt.show()
```

```python

from scipy.stats import shapiro

stat, p = shapiro(data)
print(f"Teststatistik: {stat:.3f}, p-Wert: {p:.3f}")

if p > 0.05:
    # Nullhypothese wird nicht verworfen → Daten sind normalverteilt
    print("✅ Daten sind normalverteilt")
else:
    print("❌ Daten sind nicht normalverteilt")
```
